# NLP (Natural language processing)

```python
#open text file
text_file = open(r'C:\Users\Sinam\Desktop\About_Data_Analysis.txt',encoding="utf8")
#read text file
text = text_file.read()
#datatype of the data read
print(type(text))
print("\n")
#printing text
print(text)
```

```python
print(len(text))
```

```python
import nltk
nltk.download('punkt')
from nltk import sent_tokenize
from nltk import word_tokenize
#By tokenizing the text with sent_tokenize( ), we can get the text as sentences
```

```python
sentences = sent_tokenize(text)
#the number of sentences there
print(len(sentences))
```

```python
#tokenize text with words
words = word_tokenize(text)
#the number of words
print(len(words))
```

```python
#Frequency distribution
from nltk.probability import FreqDist
#Finding frequency
fdist = FreqDist(words)
#printing 10 most common words
fdist.most_common(10)
```

```python
#Plot the graph of frequency
import matplotlib.pyplot as plt
fdist.plot(10)
```

```python
#empty list to store words:
words_no_punc = []
#removing punctuation marks: (Using the isalpha() method to separate the punctuation marks,
#along with creating a list under words_no_punc to separate words with no punctuation marks)
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
    print(words_no_punc)
    print("\n")
```

```python
print(len(words_no_punc))
```

```python
#Plotting frequency without punctuation 
fdist = FreqDist(words_no_punc)
fdist.most_common(10)
fdist.plot(10)
```
